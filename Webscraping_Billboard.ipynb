```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Output: df with 100 rows and 2 columns: name of artist and name of song
```

```python
url = "https://www.billboard.com/charts/hot-100/"
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
songs = [song.get_text() for song in soup.select("h3.c-title")]
```

```python
songs
```

```python
cleaned_songs = [songs.strip() for songs in songs]
```

```python
cleaned_songs
```

```python
# Filtering only song names
filtered_songs = [songs for songs in cleaned_songs if songs not in ["Songwriter(s):", "Producer(s):", "Imprint/Promotion Label:", 'Sabrina Carpenter Says She’s Considered Banning Phones at Her Concerts', "Gains in Weekly Performance", 'Additional Awards', '\'Star Wars\' "Looks Terrible" in Screening of Long Lost Original 1977 Version', 'Wayne Gretzky’s Daughter Paulina’s Yellow Bikini Photoshoot Shows She’s in Her ‘Unbothered’ Era', 'MLB’s Rays in Advanced Talks for $1.7B Sale to Florida Home Builder', "Man Sits On and Breaks Crystal-Encrusted 'Van Gogh' Chair in Italian Museum Before Fleeing",  '‘Dragon’s Lair’: James Bobin in Talks to Direct Video Game Adaptation for Ryan Reynolds, Netflix', 'Ye Says He Plans To Drop 10 Songs From ‘Bully’ This Week: “Still Working On The\xa0Album”', 'Just How Big Is the Sports\xa0Industry?', 'Bruce Springsteen: ‘Time Is Finite, and Your Performing Time Is Finite’','Follow Us', "Have a Tip?", 'The Daily']]
```

```python
filtered_songs
```

```python
len(filtered_songs)
```

```python
unique_songs = list(set(filtered_songs))
```

```python
len(unique_songs)
```

```python
artists = [artist.get_text() for artist in soup.select("span.c-label")]
```

```python
artists
```

```python
cleaned_artists = [artist.strip() for artist in artists]
```

```python
cleaned_artists
```

```python
# Filtering only artist names
filtered_artists = [artist for artist in cleaned_artists if not artist.isdigit() and artist not in ["NEW", "-", "RE-\nENTRY"]]
```

```python
filtered_artists
```

```python
len(filtered_artists)
```

```python
billboard_dict = {"song":unique_songs, "artist":filtered_artists}

billboard_df = pd.DataFrame(billboard_dict)
billboard_df.head()
```

```python
billboard_df
```

```python
billboard_df.to_csv("billboard_data.csv", index=False)
```
